```python
df =pd.read_csv('co.dat')
df1=pd.read_csv('1.dat')
df17= pd.read_csv('25.dat')
df25= pd.read_csv('175.dat')
```

```python
clRe17 = df[' Cl           '][:3]
cdRe17 = df['   Cd           '][:3]
cmRe17 = df['  CmPitch      '][:3]

clRe1 = df[' Cl           '][3:6]
cdRe1 = df['   Cd           '][3:6]
cmRe1 = df['  CmPitch      '][3:6]


clRe2 = df[' Cl           '][6:]
cdRe2 = df['   Cd           '][6:]
cmRe2 = df['  CmPitch      '][6:]

ang = [-2,6,14]
```

```python
xclRe17 = df17['    CL   '] 
xcdRe17 = df17['     CD   '] 
xcmRe17 = df17['    Cm         '] 

xclRe1  = df1['    CL   ']
xcdRe1  = df1['    CD   ']
xcmRe1  = df1['    Cm     ']

xclRe2  = df25['    CL   ']
xcdRe2  = df25['     CD   ']
xcmRe2  = df25['    Cm   ']

xang2    = df25['  alpha ']
xang    = df17['  alpha ']
```

```python
plt.plot(ang,clRe17,'-.o',label = 'Reynolds 1.75e6')
plt.plot(ang,clRe1,'--o',label = 'Reynolds 1e6')
plt.plot(ang,clRe2,'--o',label = 'Reynolds 2.5e5')
plt.plot(xang,xclRe17,'-.o',linewidth = 0.5,markersize = 0.5,label = 'XFOIL Reynolds 1.75e6')
plt.plot(xang,xclRe1,'--o', linewidth = 0.5,markersize = 0.5,label = 'XFOIL Reynolds 1e6')
plt.plot(xang2,xclRe2,'--o',linewidth = 0.5,markersize = 0.5,label = 'XFOIL Reynolds 2.5e5')
plt.title('Cl and AoA')
plt.xlabel('AoA')
plt.ylabel('Cl')
plt.legend()
```

```python
plt.plot(ang,cdRe17,'-.o',label = 'Reynolds 1.75e6')
plt.plot(ang,cdRe1,'--o',label = 'Reynolds 1e6')
plt.plot(ang,cdRe2,'--o',label = 'Reynolds 2.5e5')
plt.plot(xang,xcdRe17,'-.o',linewidth = 0.5,markersize = 0.5,label = 'XFOIL Reynolds 1.75e6')
plt.plot(xang,xcdRe1,'--o', linewidth = 0.5,markersize = 0.5,label = 'XFOIL Reynolds 1e6')
plt.plot(xang2,xcdRe2,'--o',linewidth = 0.5,markersize = 0.5,label = 'XFOIL Reynolds 2.5e5')
plt.title('Cd and AoA')
plt.xlabel('AoA')
plt.ylabel('Cd')
plt.legend()
```

```python
plt.plot(ang,cmRe17,'-.o',label = 'Reynolds 1.75e6')
plt.plot(ang,cmRe1,'--o',label = 'Reynolds 1e6')
plt.plot(ang,cmRe2,'--o',label = 'Reynolds 2.5e5')
plt.plot(xang,xcmRe17,'-.o',linewidth = 0.5,markersize = 0.5,label = 'XFOIL Reynolds 1.75e6')
plt.plot(xang,xcmRe1,'--o', linewidth = 0.5,markersize = 0.5,label = 'XFOIL Reynolds 1e6')
plt.plot(xang2,xcmRe2,'--o',linewidth = 0.5,markersize = 0.5,label = 'XFOIL Reynolds 2.5e5')
plt.title('Cm and AoA')
plt.xlabel('AoA')
plt.ylabel('Cm')
plt.legend()
```
